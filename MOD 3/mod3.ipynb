```python
import pandas as pd
from IPython.display import display

# ==========================================
# LOAD THE DATASETS
# ==========================================
temp_df = pd.read_csv('nyc_temperatures.csv')
weather_df = pd.read_csv('nyc_weather_2018.csv')

# ==========================================
# STEP 1: INSPECT THE DATA
# ==========================================
print("--- Initial Temperature Data Info ---")
temp_df.info()
print("\n--- Initial Weather Data Info ---")
weather_df.info()

# Check for missing values
print("\nMissing values in Temp Data:\n", temp_df.isnull().sum())
print("\nMissing values in Weather Data:\n", weather_df.isnull().sum())

# ==========================================
# STEP 2 & 3: PREPROCESSING & RATIONALE
# ==========================================
# 1. RENAMING COLUMNS
# Rationale: Renaming columns so that both datasets share a common key ('Date').
# Added: Renaming 'datatype' to 'weather_condition' to match the intended analysis.
temp_df = temp_df.rename(columns={'date': 'Date', 'temperature': 'Temperature'})
weather_df = weather_df.rename(columns={'date': 'Date', 'datatype': 'weather_condition'})

# 2. CONVERTING DATA TYPES
# Rationale: Ensuring 'Date' columns are datetime objects for accurate merging.
temp_df['Date'] = pd.to_datetime(temp_df['Date'])
weather_df['Date'] = pd.to_datetime(weather_df['Date'])

# 3. DROPPING DUPLICATES
# Rationale: Ensuring unique daily records.
temp_df = temp_df.drop_duplicates()
weather_df = weather_df.drop_duplicates()

# 4. NORMALIZING VALUES
# Rationale: Lowercasing weather codes (prcp, snow) to ensure consistency.
# This now works because we renamed the column in Step 1.
weather_df['weather_condition'] = weather_df['weather_condition'].str.lower()

# ==========================================
# STEP 4: MERGE DATA
# ==========================================
# Rationale: Inner join to keep only days where both temperature and weather codes exist.
merged_df = pd.merge(temp_df, weather_df, on='Date', how='inner')

# ==========================================
# STEP 5: VALIDATE & EXPORT
# ==========================================
print("\n--- Validation ---")
print(f"Merged Data Shape: {merged_df.shape[0]} rows, {merged_df.shape[1]} columns")
print("\nRemaining Nulls in Merged Data:\n", merged_df.isnull().sum())

print("\n--- Final Merged Dataset ---")
display(merged_df.head())

# EXPORT FOR SUBMISSION
merged_df.to_csv('Aaron_merged_weather_data.csv', index=False)
print("\nData successfully exported to Aaron_merged_weather_data.csv")
```
