These observations directly shape the preprocessing strategy in the next section.

## Data Preprocessing Techniques
Create entries to clean and blend your data sets as applicable consolidating the two or more data sets into a single collection for inclusion into your visualizations.

```python
# ---------------------------------------------------------------------------
# Step 1: Standardize column names so the three sources share a common key
# ---------------------------------------------------------------------------
education_df = education_df.rename(columns={
    "FIPS Code": "FIPS",
    "Area name": "Area_Name"
})

unemployment_df = unemployment_df.rename(columns={
    "FIPS_Code": "FIPS"
})

population_df = population_df.rename(columns={
    "FIPStxt": "FIPS"
})

# ---------------------------------------------------------------------------
# Step 2: Pivot each long-format data set into wide format
# ---------------------------------------------------------------------------
# After pivoting, each row will represent one geographic area and each
# Attribute becomes its own column.

def pivot_to_wide(df):
    wide = df.pivot_table(
        index=["FIPS", "State", "Area_Name"],
        columns="Attribute",
        values="Value",
        aggfunc="first"
    ).reset_index()
    wide.columns.name = None   # drop the leftover "Attribute" axis label
    return wide

education_wide    = pivot_to_wide(education_df)
unemployment_wide = pivot_to_wide(unemployment_df)
population_wide   = pivot_to_wide(population_df)

print(f"Education (wide):    {education_wide.shape}")
print(f"Unemployment (wide): {unemployment_wide.shape}")
print(f"Population (wide):   {population_wide.shape}")
```

```python
# ---------------------------------------------------------------------------
# Step 3: Blend the three wide-format data sets into a single table
# ---------------------------------------------------------------------------
# Outer-join on the shared keys so no area is lost if it's missing from a file.

merged_df = (
    unemployment_wide
    .merge(population_wide, on=["FIPS", "State", "Area_Name"], how="outer")
    .merge(education_wide,  on=["FIPS", "State", "Area_Name"], how="outer")
)

print(f"Final merged shape: {merged_df.shape}  (rows x columns)")
print(f"Total columns:      {len(merged_df.columns)}")
print(f"\nFirst few rows of the merged data:")
display(merged_df.head())

# Quick verification: list a sample of available columns
print("\nA sample of available columns:")
print(list(merged_df.columns)[:20], "...")
```

```python
# ---------------------------------------------------------------------------
# Step 4: Handle missing values and engineer features for the analysis
# ---------------------------------------------------------------------------

# Check missing values in the merged data
missing_summary = merged_df.isnull().sum().sort_values(ascending=False)
print("Columns with the most missing values (top 10):")
print(missing_summary[missing_summary > 0].head(10).to_string())

# Create a state-level frame (drop the U.S. national row, FIPS == 0)
states_df = merged_df[merged_df["FIPS"] != 0].copy()

# --- Engineered features ---
# Population growth from 2020 census to 2023 estimate (%)
states_df["Pop_Change_Pct_2020_2023"] = (
    (states_df["POP_ESTIMATE_2023"] - states_df["CENSUS_2020_POP"])
    / states_df["CENSUS_2020_POP"] * 100
)

# Change in unemployment rate from 2019 (pre-pandemic) to 2023
states_df["Unemployment_Change_2019_2023"] = (
    states_df["Unemployment_rate_2023"] - states_df["Unemployment_rate_2019"]
)

# Composite "Expansion Attractiveness" score for the executive audience.
# Higher bachelor's-degree %, lower unemployment, and positive population
# growth all push the score upward. Each input is min-max normalized.
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

states_df["Expansion_Score"] = (
    normalize(states_df["Percent of adults with a bachelor's degree or higher, 2019-23"])
    + normalize(-states_df["Unemployment_rate_2023"])
    + normalize(states_df["Pop_Change_Pct_2020_2023"])
) / 3

print(f"\nState-level frame shape: {states_df.shape}")
print("\nSample of new engineered columns:")
display(
    states_df[["State", "Area_Name",
               "Pop_Change_Pct_2020_2023",
               "Unemployment_Change_2019_2023",
               "Expansion_Score"]]
    .sort_values("Expansion_Score", ascending=False)
    .head(10)
)
```

### Preprocessing Summary

The preprocessing pipeline transformed three long-format data sets into a single, analysis-ready wide-format table:

1. **Standardized column names** so the three sources share a common key set (`FIPS`, `State`, `Area_Name`).
2. **Pivoted long to wide** — each unique attribute became its own column, producing one row per geographic area.
3. **Merged into a single integrated data set** using outer joins on the shared keys.
4. **Created a state-level subset** (`states_df`) by removing the U.S. national row, which would otherwise dominate any visualization scale.
5. **Engineered three features** to support the stakeholder narratives:
   - `Pop_Change_Pct_2020_2023` — population growth from the 2020 census to the 2023 estimate.
   - `Unemployment_Change_2019_2023` — change in the unemployment rate from pre-pandemic 2019 to 2023.
   - `Expansion_Score` — a normalized composite combining bachelor's-degree-or-higher percentage, low unemployment, and positive population growth. This score gives the corporate-executive audience a single, at-a-glance way to compare states.

The state-level frame is now ready for visualization.

## Data Visualizations
Create entries to visualize your data using appropriate visualization types to reveal key trends within the data collection. Integrate multiple data attributes into visualizations using appropriate techniques to derive key insights from your visualizations. In your data visualizations, you will use multi-level data modeling practices to refine your visualizations. Your visualizations should include a high-level visualization that provides a broad overview of the data for one of your two stakeholders, as well as a detailed, insight-focused visualization that highlights the key findings for the other of your two stakeholders.

```python
# ============================================================================
# AUDIENCE 1: Corporate Executives looking to expand
# ----------------------------------------------------------------------------
# High-level overview — a ranked horizontal bar chart of the top 15 states 
# by composite Expansion_Score. Bar color encodes 2022 median household 
# income, adding a market-purchasing-power dimension on top of the ranking.
# Two attributes are integrated in a single, scannable visualization.
# ============================================================================

# Build the top-15 frame
top_states = (
    states_df
    .dropna(subset=["Expansion_Score", "Median_Household_Income_2022"])
    .nlargest(15, "Expansion_Score")
    .sort_values("Expansion_Score")    # ascending so the leader appears at the top
)

fig, ax = plt.subplots(figsize=(12, 8))

# Color each bar by 2022 median household income (a separate attribute from the score)
norm = plt.Normalize(
    vmin=top_states["Median_Household_Income_2022"].min(),
    vmax=top_states["Median_Household_Income_2022"].max()
)
colors = plt.cm.YlGnBu(norm(top_states["Median_Household_Income_2022"]))

bars = ax.barh(
    top_states["Area_Name"],
    top_states["Expansion_Score"],
    color=colors, edgecolor="white", linewidth=0.6
)

# Place the numeric score at the end of each bar
for bar, score in zip(bars, top_states["Expansion_Score"]):
    ax.text(
        bar.get_width() + 0.01,
        bar.get_y() + bar.get_height() / 2,
        f"{score:.2f}",
        va="center", fontsize=10, fontweight="bold"
    )

# Colorbar to explain the color dimension
sm = plt.cm.ScalarMappable(cmap="YlGnBu", norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, shrink=0.65, pad=0.02)
cbar.set_label("Median Household Income, 2022 ($)", fontsize=10)

ax.set_xlabel("Expansion Attractiveness Score (0 = lowest, 1 = highest)", fontsize=11)
ax.set_title(
    "Top 15 States for Business Expansion (2023)\n"
    "Composite of education, low unemployment, and population growth — colored by household income",
    fontsize=14, fontweight="bold", pad=15
)
ax.set_xlim(0, top_states["Expansion_Score"].max() * 1.15)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# ============================================================================
# AUDIENCE 2: Local County Commissioners
# ----------------------------------------------------------------------------
# Detailed, multi-attribute view — a bubble scatter plotting every state 
# across four dimensions simultaneously:
#   X axis:  % adults with bachelor's degree or higher (2019-23)
#   Y axis:  unemployment rate, 2023
#   Size:    2023 population estimate
#   Color:   population change 2020-2023 (%)
# Median reference lines split the chart into quadrants for fast positioning.
# ============================================================================

plot_data = states_df.dropna(subset=[
    "Percent of adults with a bachelor's degree or higher, 2019-23",
    "Unemployment_rate_2023",
    "POP_ESTIMATE_2023",
    "Pop_Change_Pct_2020_2023"
]).copy()

fig, ax = plt.subplots(figsize=(14, 9))

scatter = ax.scatter(
    plot_data["Percent of adults with a bachelor's degree or higher, 2019-23"],
    plot_data["Unemployment_rate_2023"],
    s=plot_data["POP_ESTIMATE_2023"] / 50000,
    c=plot_data["Pop_Change_Pct_2020_2023"],
    cmap="RdYlGn",
    alpha=0.75,
    edgecolors="black", linewidth=0.6
)

# Label each bubble with the state's two-letter abbreviation
for _, row in plot_data.iterrows():
    ax.annotate(
        row["State"],
        (row["Percent of adults with a bachelor's degree or higher, 2019-23"],
         row["Unemployment_rate_2023"]),
        fontsize=8, ha="center", va="center", fontweight="bold"
    )

# Median reference lines
nat_edu   = plot_data["Percent of adults with a bachelor's degree or higher, 2019-23"].median()
nat_unemp = plot_data["Unemployment_rate_2023"].median()
ax.axvline(nat_edu,   linestyle="--", color="gray", alpha=0.5,
            label=f"Median education ({nat_edu:.1f}%)")
ax.axhline(nat_unemp, linestyle="--", color="gray", alpha=0.5,
            label=f"Median unemployment ({nat_unemp:.1f}%)")

# Quadrant annotations
ax.text(0.98, 0.97, "High education,\nhigher unemployment",
        transform=ax.transAxes, ha="right", va="top",
        fontsize=9, alpha=0.55, style="italic")
ax.text(0.02, 0.97, "Lower education,\nhigher unemployment",
        transform=ax.transAxes, ha="left", va="top",
        fontsize=9, alpha=0.55, style="italic")
ax.text(0.98, 0.03, "★ Target zone:\nhigh education,\nlow unemployment",
        transform=ax.transAxes, ha="right", va="bottom",
        fontsize=9, alpha=0.75, style="italic", fontweight="bold")
ax.text(0.02, 0.03, "Lower education,\nlow unemployment",
        transform=ax.transAxes, ha="left", va="bottom",
        fontsize=9, alpha=0.55, style="italic")

# Colorbar for the growth dimension
cbar = plt.colorbar(scatter, ax=ax, shrink=0.75, pad=0.02)
cbar.set_label("Population change 2020–2023 (%)", fontsize=10)

ax.set_xlabel("% Adults with Bachelor's Degree or Higher (2019–23)", fontsize=11)
ax.set_ylabel("Unemployment Rate 2023 (%)", fontsize=11)
ax.set_title(
    "Workforce Landscape: State-by-State Detail (2023)\n"
    "Bubble size = state population; color = population growth since 2020",
    fontsize=14, fontweight="bold", pad=15
)
ax.legend(loc="upper left", fontsize=9, framealpha=0.9)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```
